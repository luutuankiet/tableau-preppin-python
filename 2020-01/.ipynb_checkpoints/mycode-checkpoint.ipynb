```python
# setup
import pandas as pd
import os
import numpy as np
import time
current_dir = os.getcwd()
sub_dir = 'inputs'
files = os.path.join(current_dir, sub_dir)
csv_file = [file for file in os.listdir(files) if file.endswith('.csv')]
csv_file
```

```python
for csv in csv_file:
    csv_path = os.path.join(current_dir,sub_dir, csv)
    df = pd.read_csv(csv_path)
df.head(3)

# get the levels using isnumeric
levels = []
for item in df['Item']:
    level = sum(
        [i.isnumeric() for i in item]
    )
    levels.append(level)
df['Level'] = levels



# create hierrachy : define parents and children 
# 3 as the last children, 1 as the first parent
parent_1 = (df['Level'] == 1) 
parent_2 = (df['Level'] == 2)
parent_3 = (df['Level'] == 3)
children_1 =  (df['Level'] == 3) 
children_2 =  (df['Level'] == 2) 



# string slicing : use str[0]
# indicate the subset : specific rows matching "children" condition 
df.loc[children_1,'Parent'] = df['Item'].str[:3]
df.loc[children_2,'Parent'] = df['Item'].str[0]


df.loc[parent_1,'ID'] = df['Item'].str[0]
df.loc[parent_2,'ID'] = df['Item'].str[:3]
df.loc[parent_3,'ID'] = df['Item'].str[:5]


def assign_df(df, id_key):
    df.loc[df['ID'] ==  id_key ,'Profit'] = df.loc[df['Parent'] == id_key,'Profit'].sum()

for index, row in df.iterrows():
    level_key = row['Level']
    parent_key = row['Parent']
    id_key = row['ID']
    # print(f'processing level {level_key}, parent {parent_key}, id {id_key}',end='\r')
    # time.sleep(1) 
    print(f'processing level {level_key}, parent {parent_key}, id {id_key}',end='\r')


    # ignore calcs for child
    
    if level_key == 3:
        continue

    # # 1st sequence for n+1 parents
    if level_key == 2:
        assign_df(df,id_key)

    first_seq_ran = True
    
    time.sleep(1) 

for index, row in df.iterrows():
    level_key = row['Level']
    parent_key = row['Parent']
    id_key = row['ID']    
    # # 2nd sequence for upper parents
    if level_key == 1 and first_seq_ran:
        # df.loc[df['ID'] ==  id_key ,'Profit'] = df.loc[df['Parent'] == id_key,'Profit'].sum()
        assign_df(df,id_key)
df





```
