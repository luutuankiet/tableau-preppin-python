```python
#create parent category
raw_df["ParentID"] = np.nan

# indicate outer parent and assign it to 1 level below, ie 2

outer_children = raw_df['level'] == 2
raw_df.loc[outer_children,'ParentID'] = raw_df.loc[outer_children,'Item'].str[0]


# # indicate inner parent and assign it to level 3 the children
inner_children = raw_df['level'] == 3
raw_df.loc[inner_children,'ParentID'] = raw_df.loc[inner_children,'Item'].str[:3]
raw_df

```

```python
raw_df['Profit'] = raw_df.groupby('ParentID')['Profit'].transform(lambda x: x.sum() if x.name != '3' else x)
raw_df
```
